AIM #2: Finding issues (empty, NAs, incorrect value, incorrect format, outliers, etc.) 
1. Find out how many missing values there are in the dataset
2. For the 'Age' column, find the best way to handle the missing values
    2.1. Use an appropriate plot to study the nature of the 'Age' column
    2.2. Figure out what is the best way to calculate the central tendency of the 'Age' column based on the above plot
    2.3. Using the most suitable central tendency measure, fill the missing values in the age column
3. Decide what is the best way to handle the missing values in the 'Cabin' columns
4. Similarly, decide what is the best way to handle the missing values in the 'Embarked' columns
5. Handle the incorrect data under the 'Survived' columns using appropriate measure
6. Handle the incorrectly formatted data under the 'Fare' column


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic_df = pd.read_csv('titanic.csv')

missing_values_count = titanic_df.isnull().sum()
print("Number of missing values:")
print(missing_values_count)

sns.histplot(titanic_df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()

age_median = titanic_df['Age'].median()

titanic_df['Age'].fillna(age_median, inplace=True)

titanic_df['HasCabin'] = titanic_df['Cabin'].notnull().astype(int)
titanic_df.drop('Cabin', axis=1, inplace=True)

embarked_values = titanic_df['Embarked'].unique()
print("Values in the 'Embarked' column:", embarked_values)
embarked_mode = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(embarked_mode, inplace=True)

survived_values = titanic_df['Survived'].unique()
if len(survived_values) > 2 or (survived_values!= 0).any() and (survived_values!= 1).any():
    print("There are abnormal values in the 'Survived' column and need further inspection and handling")

    titanic_df = titanic_df[(titanic_df['Survived'] == 0) | (titanic_df['Survived'] == 1)]

print("Data type of the 'Fare' column:", titanic_df['Fare'].dtype)

if titanic_df['Fare'].dtype == object:
    titanic_df['Fare'] = pd.to_numeric(titanic_df['Fare'], errors='coerce')

titanic_df = titanic_df[titanic_df['Fare'] > 0 & (titanic_df['Fare'] <= 1000)]
```

AIM #3: Grouping 
1. Find out the average fare grouped by Pclass
    1.1. Plot the above using a suitable plot
2. Find out the average fare grouped by Sex
    2.1. Plot the above using a suitable plot

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_df = pd.read_csv('titanic.csv')

print(titanic_df.dtypes)

titanic_df['Fare'] = pd.to_numeric(titanic_df['Fare'], errors='coerce')

titanic_df.dropna(subset=['Fare'], inplace=True)

average_fare_by_pclass = titanic_df.groupby('Pclass')['Fare'].mean()

plt.figure(figsize=(8, 6))
average_fare_by_pclass.plot(kind='bar', color='skyblue')
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()

average_fare_by_sex = titanic_df.groupby('Sex')['Fare'].mean()

plt.figure(figsize=(8, 6))
average_fare_by_sex.plot(kind='bar', color='salmon')
plt.title('Average Fare by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Fare')
plt.show()
```

AIM #4: Dataset visualization using pandas

1. Plot the distribution of 'Age' using a suitable plot
2. Plot the distribution of 'Fare' using a suitable plot
3. Plot the distribution of 'Pclass' using a suitable plot
4. Plot the distribution of 'Survived' using a suitable plot
5. Plot the distribution of 'Embarked' using a suitable plot
6. Plot the distribution of 'Fare' grouped by 'Survived'
7. Plot the distribution of 'Fare' grouped by 'Pclass'
8. Plot the distribution of 'Age' grouped by 'Survived'
9. Plot the distribution of 'Age' grouped by 'PClass'

11. Combine the 'SibSp' and 'Parch' and plot its distribution grouped by 'Pclass'

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic_df = pd.read_csv('titanic.csv')

titanic_df['Fare'] = pd.to_numeric(titanic_df['Fare'], errors='coerce')


plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Fare'].dropna(), kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
titanic_df['Pclass'].value_counts().plot(kind='bar')
plt.title('Distribution of Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
titanic_df['Survived'].value_counts().plot(kind='bar')
plt.title('Distribution of Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
titanic_df['Embarked'].value_counts().plot(kind='bar')
plt.title('Distribution of Embarkation Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=titanic_df)
plt.title('Distribution of Fare grouped by Survived')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=titanic_df)
plt.title('Distribution of Fare grouped by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=titanic_df)
plt.title('Distribution of Age grouped by Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic_df)
plt.title('Distribution of Age grouped by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='FamilySize', data=titanic_df)
plt.title('Distribution of Family Size grouped by Survived')
plt.xlabel('Survived')
plt.ylabel('Family Size')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='FamilySize', data=titanic_df)
plt.title('Distribution of Family Size grouped by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Family Size')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic_df)
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

correlation_matrix = titanic_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=titanic_df)
plt.title('Relationship between Age and Fare by Sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

AIM #5: Correlation

1. Generate a correlation matrix for the entire dataset
2. Find correlation between 'Age' and 'Fare'
3. What other possible correlations can be found in the dataset?

```python
import pandas as pd

titanic_df = pd.read_csv('titanic.csv')

print("Unique values in 'Age':", titanic_df['Age'].unique())
print("Unique values in 'Survived':", titanic_df['Survived'].unique())
print("Unique values in 'Pclass':", titanic_df['Pclass'].unique())
print("Unique values in 'Fare':", titanic_df['Fare'].unique())

titanic_df['Fare'] = pd.to_numeric(titanic_df['Fare'], errors='coerce')

correlation_matrix = titanic_df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

age_fare_correlation = titanic_df['Age'].corr(titanic_df['Fare'])
print(f"\nCorrelation between 'Age' and 'Fare': {age_fare_correlation}")

titanic_df['Survived'] = pd.to_numeric(titanic_df['Survived'], errors='coerce')

survived_age_correlation = titanic_df['Survived'].corr(titanic_df['Age'])
print(f"Correlation between 'Survived' and 'Age': {survived_age_correlation}")

survived_fare_correlation = titanic_df['Survived'].corr(titanic_df['Fare'])
print(f"Correlation between 'Survived' and 'Fare': {survived_fare_correlation}")

pclass_age_correlation = titanic_df['Pclass'].corr(titanic_df['Age'])
print(f"Correlation between 'Pclass' and 'Age': {pclass_age_correlation}")

pclass_fare_correlation = titanic_df['Pclass'].corr(titanic_df['Fare'])
print(f"Correlation between 'Pclass' and 'Fare': {pclass_fare_correlation}")
```
